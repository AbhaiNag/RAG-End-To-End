```python
# Load the documents

from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader

loader = DirectoryLoader('./sample_data/', glob="*.txt", show_progress=True, loader_cls=TextLoader)

kb_docs = loader.load()
```

```python
# Chunk the loaded documents

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

chunks = text_splitter.split_documents(kb_docs)
```

```python
print("Number of Documents:", len(kb_docs))
print()
print("Number of Chunks:", len(chunks))
```

```python
# Step 2: Create a Retriever Object

retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})
```

```python
# Step 3: Initialize a Chat Prompt Template

from langchain_core.prompts import ChatPromptTemplate

PROMPT_TEMPLATE = """
Answer the question based only on the following context:
{context}
Answer the question based on the above context: {question}. Provide a concise, one-liner answer.
Don’t justify your answers.
Don’t extrapolate information.
Don’t give information not mentioned in the CONTEXT INFORMATION.
Do not say "according to the context" or "mentioned in the context" or similar.
"""

prompt_template = ChatPromptTemplate(
    messages=[
        PROMPT_TEMPLATE
    ]
)
```

```python
# Step 4: Initialize a Generator (i.e. Chat Model)

from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(api_key='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')
```

```python
# Step 5: Initialize a Output Parser

from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
generator_chain = prompt_template | chat_model | parser
```

```python
# Helper function to join the retrieved chunks

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
# Step 6: Define a RAG Chain

from langchain_core.runnables import RunnablePassthrough

rag_chain = {
    "context": retriever | format_docs,
    "question": RunnablePassthrough()
} | generator_chain
```
